```python
import os
import pandas as pd

DATADIR = "A_TXT"
DATAFILE = "1Amar.txt"

datafile = os.path.join(DATADIR, DATAFILE)
df = pd.read_csv(datafile, sep="	", skiprows=7, header=None, names=['RF', 'BF', 'VM', 'ES', 'FE'])

df = df.dropna()

df.tail()
```

```python
%matplotlib notebook
import matplotlib.pyplot as plt

plt.figure()
plt.subplot(1, 2, 1)
plt.plot(df['RF'], linewidth=0.2)

plt.subplot(1, 2, 2)
plt.plot(df['BF'], linewidth=0.2)
```

```python
%matplotlib notebook
import matplotlib.pyplot as plt

plt.figure()
ax1 = plt.subplot(2, 1, 1)
plt.plot(df['RF'], linewidth=0.2)
plt.xlabel('Amostra')
plt.ylabel('mV').set_rotation(0)
plt.title('Recto Femural')

    
    
# pass sharey=ax1 to ensure the two subplots share the same y axis
ax2 = plt.subplot(2, 1, 2, sharey=ax1).yaxis.set_visible(False)
plt.plot(df['BF'], linewidth=0.2)
plt.xlabel('Amostra')
#plt.ylabel('mV')
plt.title('Femoral Biceps')
```

```python
plt.figure()
ax3 = plt.subplot(1, 2, 1)
plt.plot(df['VM'], linewidth=0.2)
plt.xlabel('Amostra')
plt.ylabel('mV').set_rotation(0)
plt.title('Vastus Medialis')
# pass sharey=ax1 to ensure the two subplots share the same y axis
ax4 = plt.subplot(1, 2, 2, sharey=ax3).yaxis.set_visible(False)
plt.plot(df['ES'], linewidth=0.2)
plt.xlabel('Amostra')
#plt.ylabel('mV')
plt.title('Smitendinosus')
```

```python
plt.figure()
ax5 = plt.plot
plt.plot(df['FE'], linewidth=0.5)
plt.xlabel('Amostra')
plt.ylabel('Deg').set_rotation(0)
plt.title('Flexion at the knee')
```
