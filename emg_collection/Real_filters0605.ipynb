```python
import matplotlib.pyplot as plt 
import numpy as np
%matplotlib inline
```

## Filtro Passa-alta 20Hz (15.9Hz)

```python
# filtra em cerca de 15,9Hz
entrada = np.array([1., 1., 1.,1., 1.,1.,1.,1.,1.])
saida = np.array([0.164, 0.384,0.496, 0.720,0.76,0.864, 0.968, 0.984, 1.])
freq =np.array([1.26, 4.99,8.00,14.98,16.03, 25.15,38.01, 50.05, 79.58])
p = 20*np.log10(np.divide(saida, entrada))

#frenquencia vs saida
plt.plot(freq, saida, "*--")
plt.plot(freq, entrada,  c='r')
plt.title("Tensao de saida de acordo com a frequencia, entrada em vermelho")
plt.xlabel("Frequencia")
plt.ylabel("Vout")
plt.ylim([0, 1.1])
plt.grid()
plt.show()

# Potencia da saida
plt.plot(freq, p)
plt.title("Intensidade da saida em dB")
plt.xlabel("Frequencia")
plt.ylabel("dB")
plt.ylim([-18, 3])
plt.grid()
plt.show()
```
